```python
# prompt: Create an app for summarization using gradio

!pip install gradio

import gradio as gr
from transformers import pipeline

summarizer = pipeline("summarization")

def summarize(text):
  result = summarizer(text, min_length=5, max_length=20)
  return result[0]['summary_text']

iface = gr.Interface(
    fn=summarize,
    inputs="text",
    outputs="text",
    title="Text Summarizer",
    description="Summarize any text using the Hugging Face Transformers library.",
)

iface.launch()
```
